```python
df['Spending Score (1-100)'].hist()
```

```python
data = df['Spending Score (1-100)'].value_counts()

top_n = 10
data_top = data.head(top_n)

plt.pie(data_top.values, labels=data_top.index, autopct='%1.1f%%', startangle=140)
plt.title(f"-{top_n}  'Spending Score (1-100)'")
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
y_pred = KMeans(n_clusters=5, random_state=1, n_init='auto').fit_predict(df)
plt.scatter(list(range(df['Spending Score (1-100)'].shape[0])), df['Spending Score (1-100)'], c=y_pred)
plt.title("  ")
plt.show()
```
